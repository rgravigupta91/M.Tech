# EDA: Bivariate Analysis

```python
tips = sns.load_dataset('tips')
titanic = pd.read_csv('Titanic-Dataset.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.head()
```

```python
titanic.head()
```

```python
flights.head()
```

```python
iris.head()
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(tips,x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(tips,x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(titanic,x=titanic['Pclass'],y=titanic['Fare'])
```

```python
sns.barplot(titanic,x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

## 5. Heatmap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
```

```python
titanic.groupby('Pclass')['Survived'].mean()
```

## 6. Cluster plot

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical-Numerical)

```python
flights.head()
```

```python
flights1 = flights.groupby('year')['passengers'].sum().reset_index()
flights1
```

```python
sns.lineplot(x=flights1['year'],y=flights1['passengers'])
```

```python
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
