# Stack, Unstack

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(23)
n84 = np.random.randint(low=10, high=100, size=(8,4))
n84
```

```python
row_labels_level0 = ['TATA','BIRLA','Wipro','ITC']
row_labels_level1 = ['FMCG','TECH']

col_labels_level0 = [2022,2023]
col_labels_level1 = ['Q1','Q2']
```

```python
row_labels = pd.MultiIndex.from_product([row_labels_level0, row_labels_level1])
col_labels = pd.MultiIndex.from_product([col_labels_level0, col_labels_level1])
```

```python
df1 = pd.DataFrame(n84, index=row_labels, columns=col_labels)
```

```python
df1.index
```

```python
df1.columns
```

```python
df1.stack()
```

```python
df1
```

```python
df1.stack(level=0)
```

```python
df1.stack(level=0).index
```

```python
df1.stack(level=0).columns
```

```python
df1.stack(level=0).unstack(level=0)
```

```python
df1.stack(level=0).unstack(level=0).unstack()
```

```python
df1
```

```python
df1
```

```python
df1.unstack(1).stack(0).unstack().stack(0).unstack(0).stack()
```

# Concatenation

```python
df11 = pd.read_csv('./Data/lds11.csv')
df12 = pd.read_csv('./Data/lds12.csv')
df13 = pd.read_csv('./Data/lds13.csv')
```

```python
print(df11)
print(df12)
print(df13)
```

```python
pd.concat([df11,df12])
```

```python
pd.concat([df11,df12], axis=1)
```

```python
pd.concat([df11,df12], axis=1, join='outer')
```

```python
pd.concat([df11,df12], axis=1, join='outer', keys=['f11','f12'])
```

```python
pd.concat([df11,df12], axis=0, join='outer', keys=['f11','f12'])
```

```python
pd.concat([df11,df12], join='inner', keys=['f11','f12'])
```

```python
pd.concat([df11,df12], axis=1, join='inner', keys=['f11','f12'])
```

# Merge

```python
pd.merge(df21,df22,how='inner')
```

```python
pd.merge(df21,df22,how='inner', left_on='stock_id', right_on='stock_name')
```

```python
pd.merge(df21,df22,how='outer', left_on='stock_id', right_on='stock_name')
```
