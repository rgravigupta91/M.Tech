## 5. Create a function that should be able to create a new feature(Variable) called season using the month column.

```python
def addSeason(df, mon_col):
    season_dic = {'jan':'winter', 'feb':'winter', 'mar':'spring', 'apr':'spring', 'may':'spring', 'jun':'summer',
                  'jul':'summer', 'aug':'summer', 'sep':'autumn', 'oct':'autumn', 'nov':'autumn', 'dec':'winter'}
    df['season'] = df[mon_col].map(season_dic)

addSeason(bank_df, 'month')
bank_df
```

## 13. Use the violin plot and also the box plot to find the distribution of the balance for each class of the Y column. And try to tell why we have a Violin plot and Box plot both rather than one.

```python
plt.figure(figsize=(15,9), dpi=100)
sns.violinplot(x=bank_df.balance, y=bank_df.y)
plt.show()
```

```python
plt.figure(figsize=(15,9), dpi=100)
sns.boxplot(x=bank_df.balance, y=bank_df.y)
plt.show()
```

### Boxplot vs Violin Plot
The box plot shows the distribution of numerical values, which include the median, quartiles and outliers. The box plot also hides the individual data points.
The violin plot is called as such because the plot looks like a violin. The plot is similar to box plot but it also adds a density plot. The violin helps visualise the dataset's distribution and shows the density at different values.

## 15. Use Box plot and strip plot to know the distribution of the Pdays with respect to Y classes and differentiate both plots.

```python
plt.figure(figsize=(15,9), dpi=100)
sns.boxplot(x=bank_df.pdays, y=bank_df.y)
plt.show()
```

```python
plt.figure(figsize=(15,9), dpi=100)
sns.stripplot(x=bank_df.pdays, y=bank_df.y)
plt.show()
```

The box plot only shows the distribution of the campaigns to subscribed and non-subscribed customers.
But it is the strip plot that shows that there the subscribed customers are contacted in intervals close to 100 days, while the non subcribed customers are contacted more often.
