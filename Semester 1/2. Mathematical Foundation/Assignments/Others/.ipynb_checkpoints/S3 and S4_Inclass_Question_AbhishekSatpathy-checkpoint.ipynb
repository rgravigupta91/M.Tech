#### Q3. Show horizontally One-third of the image in Red next one-third in blue and the last one third in green

```python
one_third_width = width//3
```

```python
image_array = np.array(image)

# dimension of image
height, width, channels = image_array.shape

image_array[:, :one_third_width] = [255, 0, 0]  # Red
image_array[:, one_third_width:2*one_third_width] = [0, 0, 255]  # Blue
image_array[:, 2*one_third_width:] = [0, 255, 0]  # Green

modified_image = Image.fromarray(image_array)

modified_image
```

#### Q4. Scale down the image by half

```python
new_width = width // 2
new_height = height // 2

scaled_image = image.resize((new_width, new_height))

scaled_image
```

### Session 4

#### Q9. Read the data S3_1.

#### Q10. Print out the number of features in the dataset.

#### Q11. Print out the amount of variation we loose with the reduction of 10 feature from the dataset.

#### Q12. Print out the maximum numbers of feature we can lay off and still capture 95% of the total variation.

```python
#Q9
data = pd.read_excel("S3_1.xlsx")
data
```

```python
#Q10
num_features = data.shape[1]

print("Number of features in the dataset:", num_features)
```

```python
#Q11
original_variance = data.var().sum()
reduced_dataframe  = data.drop(data.columns[9], axis=1)
new_variance = reduced_dataframe.var().sum()

print(original_variance-new_variance)
```

```python
#Q12
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(data)

explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

n_components_95 = np.argmax(cumulative_variance_ratio >= 0.95) + 1
n_components_95
```
