## Question 1 : Function

A dam releases 2M gallons of water on a month. Then increases the water release exponentially for next 3 months on every 5th month the government closes the dam for a month and again start releasing water in the same way.

1. Construct the funtion for the amount of water release from thte dam.
2. Draw the function of the monthly amount of water release from the dam. for 4 years .

### 1.1 : Construct the funtion for the amount of water release from the dam.

```python
import numpy as np
index = np.arange(1,4*12+1)
ind = index % 5
print(ind)
```

```python
def water(ind):
    mile_t = 2 ** ind 
    return mile_t
water(ind)
```

### 1.2 : Draw the function of the monthly amount of water release from the dam for 4 years .

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.scatter(index,water(ind),color = 'r')
plt.plot(index,water(ind))
plt.xlabel('Number of the month')
plt.ylabel('Amount of water released')
plt.xticks(index)
plt.ylim(0,22)
plt.show()
```

### 3.4 : Find the rate of change of movement of the partical at 90 seconds

### 3.4.1 : Use the limit concept.

```python
import math
lim = sym.limit((((2 * ((x+h) ** 2)) - (3 * (x+h)) + 1 ) -  ((2 * (x ** 2)) - (3 * x) + 1))/h , h, 0)
limx = sym.lambdify(x , lim)
#rad90 = math.radians(90)
limx(90)
```

### 3.4.2 : use the differential fuction

```python
x = sym.Symbol('x')
def rate_ch_dif_sp(t):
    dif = sym.diff(((2 * (x ** 2)) - (3 * x) + 1),x,1)
    dx = sym.lambdify(x , dif)
    return dx(t)
rate_ch_dif_sp(90)
```

### 3.5 : Draw the rate of change of movement of the partical  at 90 seconds

```python
#rad = np.array(list(map(lambda x :math.radians(x),time)))
rate_c = np.array(list(map(lambda x :rate_ch_dif_sp(x),time)))
```

## Question 5 : Application

A garment company manufactures 100 rills of cloth per day.
Each standard rill costs Rs. 20 and the company charges Rs 50 to its customers.
After 5 months the company buys a new equipment for Rs 30,000 and decides to pay in EMI of Rs 3000.
So the company include that cost as regular cost in the revenue of each month.

1. Draw the graph of 
    a. monthly Profit, 
    b. monthly expense and 
    c. monthly earning 
    of the Company for first 3 years of it's business.

2. Find the rate of change in profit

### 5.1.a : Draw the graph of monthly earning of the Company for first 3 years of it's business.

```python
rills_day = 100
rills_monthly = 30 * 100
cost = 20 
price = 50
index = np.arange(1,rills_day+1)
```

```python
def Price(x):
    price= 50 * rills_monthly
    return price

pr = np.array(list(map(lambda x : Price(x),index)))
```

```python
plt.plot(index , pr)
plt.xlabel('Month')
plt.ylabel('Sales amount')
plt.show()
```

### 5.1.b : Draw the graph of monthly expense of the Company for first 3 years of it's business.

```python
def Cost(x):
    if x < 5: ##Day company gives the order of raw material
        cost= 20 * rills_monthly
    elif (x < int(30000/3000) +1) and (x >5):
        cost= 20 * rills_monthly + 3000
    else:
        cost= 20 * rills_monthly
    return cost
cos = np.array(list(map(lambda x : Cost(x),index)))
```

```python
plt.plot(index , cos)
plt.xlabel('Month')
plt.ylabel('Cost')
plt.show()
```

### 5.1.c: Draw the graph of monthly Profit of the Company for first 3 years of it's business.

```python
def Profit(x):
    profit = Price(x) - Cost(x)
    return profit
pro = np.array(list(map(lambda x : Profit(x),index)))
```

```python
plt.plot(index , pro)
plt.xlabel('Month')
plt.ylabel('Profit')
plt.show()
```

### 5.2: Find the rate of change in profit

```python
Change_in_Profit = [Profit(x)-Profit(x-1) for x in index[1:]]
```

```python
plt.figure(figsize =(20,5))
plt.plot(index[1:] , Change_in_Profit)
plt.xlabel('Month')
plt.ylabel('Change in Profit')
plt.show()
```
