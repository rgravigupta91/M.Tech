# Question 1 : Optimizing function with one variable

A particle travels in a particular path.The function of the path could be written as (3 * cos(1/(5 + t))) + (1/((1 + t) ** 2)) + (7 * (t ** 5))
where t reffers the time in second.(110 sec)

1. Draw the path of the partical

2. Find out the velocity of the partical

3. Draw the velocity of the partical

4. Find out the time(in sec) where the partical moves the fastest

5.  Find out the time(in sec) where the partical moves the slowest

6. Find out if there is any time point where the partical does not move

### 1.1. Draw the path of the partical

```python
import numpy as np
import matplotlib.pyplot as plt
time = np.arange(1,110)
travel = (3 * np.cos(1/(5 + time))) + (1/((1 + time) ** 2)) + (7 * (time ** 5))
plt.figure(figsize = (20,8))
plt.plot(time , travel)
plt.xlabel('Time')
plt.ylabel('Movement')
plt.show()
```

### 1.2. Find out the velocity of the partical

```python
import sympy as sym
t = sym.Symbol('t')
vel = sym.diff((3 * sym.cos(1/(5 + t))) + (1/((1 + t) ** 2)) + (7 * (t ** 5)),t)
vel
```

```python
velocity = sym.lambdify(t , vel)
velocity(time)
```

### 1.3. Draw the velocity of the partical

```python
plt.figure(figsize = (20,8))
plt.plot(time,velocity(time))
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.show()
```

### 1.4. Find out the time(in sec) where the partical moves the fastest

```python
def fast(x):
    return -velocity(x)
```

```python
from scipy.optimize import minimize_scalar
res = minimize_scalar(fast , bounds=(1, 119), method='bounded')
res
```

```python
-fast(res.x.round(0))
```

### 1.5.  Find out the time(in sec) where the partical moves the slowest

```python
def slow(x):
    return velocity(x)
```

```python
from scipy.optimize import minimize_scalar
res = minimize_scalar(slow , bounds=(1, 119), method='bounded')
res
```

```python
slow(res.x.round(0))
```

```python
time[velocity(time)== 0]
```

### 1.6. Find out if there is any time point where the partical does not move

```python
from scipy.optimize import root
sol = root(velocity, 0.3)
print (sol)
```

### 2.3. Constract a revenue function  with parameters as number of units of product the company produces of each Product-1 and Product-2 and time

```python
def Revenue(a,t):
    func = (a * ( 48 - (1 / (5 + (t**2))) + 10 * np.cos(np.pi * t / 180))) + (((5000 - (a*50))/50) * (52 - t + ( 10 * np.tan(np.pi * t / 180))))
    return func
```

### 2.5. Find out the optimum split of number of unit produced of Product-1 & Product-2 and the best time point

```python
def loss(arguments):
    a,t = arguments
    pr = ( 5000 - (a * ( 48 - (1 / (5 + (t**2))) + 10 * np.cos(np.pi * t / 180))) - (((5000 - (a*50))/50) * (52 - t + ( 10 * np.tan(np.pi * t / 180)))))
    return pr
```

```python
from scipy.optimize import minimize

initial_guess = [0, 80]

res = minimize(loss,initial_guess,method='SLSQP',bounds = [[0,100],[1,900]],options={'xtol': 1e-20, 'disp': True,'ftol':2} )#,method='Nelder-Mead')
res.x
```

### 2.6. Find out the optimum revenue earned

```python
Revenue(res.x[0].round(0),res.x[1].round(0))
```

### 4.4 Prepare the cost function

```python
def Cost(a,b):
    cost = sym.square(Y - (a + b * x))/N
    return cost
```

### 4.5 Find the derivative of cost with respect to 'a'.

```python
def Cost_der_a(a,b,Y,x):   
    return -(((Y - (a + b * x))/N).sum()
```

### 4.6 Find the value of the derivative in 4.5 with the initial values of 'a' & 'b' and store it in a variable A

### 4.7 Find the derivative of cost with respect to 'b'.

```python
def Cost_der_b(a,b,Y,x):   
    return -(((Y - (a + b * x))*(x))/N).sum()
```

### 4.8 Find the value of the derivative in 4.7 with the initial values of 'a' & 'b' and store it in a variable B

```python
B = Cost_der_b(a_in,b_in,Y,x)
```

```python
b_in - (learning_rate * B) 
```

### 4.12 if the difference in smaller than error rate stop otherwise recalculate the value of 4.6 & 4.8 with the new pair of parameter values

```python
i = 0
if error_a > error_rate :
    A = Cost_der_a(list_a[i+1],list_b[i+1],Y,x)
    B = Cost_der_b(list_a[i+1],list_b[i+1],Y,x)
else:
    break
```

### 4.13 Run a loop to repeat 4.9, 4.10, 4.11, 4.12 until we reach to the stopping condition.

```python
i = 0
if error_a > error_rate or error_b > error_rate:
    a_in -=  (learning_rate * A) 
    b_in -=  (learning_rate * B) 
    list_a.append(list_a[i+1] -(learning_rate * A) )
    list_b.append(list_b[i+1] -(learning_rate * B) )
    error_a = list_a[1] - list_a[0]
    error_b = list_b[1] - list_b[0]
    A = Cost_der_a(list_a[i+1],list_b[i+1],Y,x)
    B = Cost_der_b(list_a[i+1],list_b[i+1],Y,x) 
    error_a = list_a[i+1] - list_a[i]
    error_b = list_b[i+1] - list_b[i]
else:
    break
```
