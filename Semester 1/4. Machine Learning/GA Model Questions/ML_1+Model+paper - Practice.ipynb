```python
#Splitting company name from CarName column
CompanyName = df['CarName'].apply(lambda x : x.split(' ')[0])
df.insert(3,"CompanyName",CompanyName)
df.drop(['CarName'],axis=1,inplace=True)
df.head()
```

##### Fixing invalid values
- There seems to be some spelling error in the CompanyName column.

    - `maxda` = `mazda`
    - `Nissan` = `nissan`
    - `porsche` = `porcshce`
    - `toyota` = `toyouta`
    - `vokswagen` = `volkswagen` =  `vw`

```python
df.CompanyName = df.CompanyName.str.lower()

def replace_name(a,b):
    df.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

df.CompanyName.unique()
```

```python
def histogram_plot(attr,df,sequence=True):
    plt.figure(figsize=(20,8))

    plt.subplot(1,2,1)
    plt.title(attr+' Histogram')
    if sequence == True:
        sns.countplot(x=df[attr], palette=("cubehelix"))
    else:
        plt1 = df[attr].value_counts().plot(kind='bar')

    plt.subplot(1,2,2)
    plt.title(attr+' vs Price')
    sns.boxplot(x=df[attr], y=df.price, palette=("cubehelix"))
    
#    plt.subplot(2,2,3)
    df1 = pd.DataFrame(df.groupby([attr])['price'].mean().sort_values(ascending = False))
    df1.plot.bar(figsize=(6,6))
    plt.title(attr+' vs Average Price')

    plt.show()
```

#### Feature Engineering

```python
#Fuel economy
df['fueleconomy'] = (0.55 * df['citympg']) + (0.45 * df['highwaympg'])
```

```python
#Binning the Car Companies based on avg prices of each Company.
df['price'] = df['price'].astype('int')
temp = df.copy()
table = temp.groupby(['CompanyName'])['price'].mean()
temp = temp.merge(table.reset_index(), how='left',on='CompanyName')
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
df['carsrange'] = pd.cut(temp['price_y'],bins,right=False,labels=cars_bin)
df.head()
```

#### Bivariate Analysis

```python
plt.figure(figsize=(8,6))

plt.title('Fuel economy vs Price')
sns.scatterplot(x=df['fueleconomy'],y=df['price'],hue=df['drivewheel'])
plt.xlabel('Fuel Economy')
plt.ylabel('Price')

plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(25, 6))

df1 = pd.DataFrame(df.groupby(['fuelsystem','drivewheel','carsrange'])['price'].mean().unstack(fill_value=0))
df1.plot.bar()
plt.title('Car Range vs Average Price')
plt.show()
```
