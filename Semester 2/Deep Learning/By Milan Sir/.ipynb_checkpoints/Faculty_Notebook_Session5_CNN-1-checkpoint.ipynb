# Image Augumentation

1. **Increases Dataset Size**: Image augmentation artificially expands the size of the training dataset by generating new, varied images, which helps in improving the model's generalization without the need to collect more data.

2. **Reduces Overfitting**: By creating slightly different variations of the same image (through flipping, rotation, zooming, etc.), augmentation helps the model to generalize better and avoid overfitting on the training data.

3. **Improves Model Robustness**: Augmentation introduces distortions and variations that simulate real-world scenarios, enabling the model to perform better on unseen and varied test data.


```python
# create few more images by stretching, zooming, shrinking, rotating, flipping the available images
#This will make the model to learn very generically about the data
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen=ImageDataGenerator(rescale=1/255.,
        rotation_range=45,     #Random rotation between 0 and 45
        width_shift_range=0.2,   #% shift
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='reflect')
```

```python
test_datagen=ImageDataGenerator(rescale=1/255.) # We dont experiment the test data generally
# only scaling is enough for test data
```

```python
# Laod the images from the train and test folders by resizing it for 128x128
#Image augumentaion defined above will be performed for each batch of images 
# class mode is "categorical" for multiclass problem ("binary" for binomial problem) 
training_set=train_datagen.flow_from_directory('Training',
                                              target_size=(128,128),
                                              batch_size=128,
                                              class_mode='categorical')

# Total of 2736 images corresponds to 5 differnt classses are loaded in to this notebook 
                                                   #from the training folder (harddrive)
```

```python
test_set=test_datagen.flow_from_directory('Testing',
                                              target_size=(128,128),
                                              batch_size=128,
                                              class_mode='categorical')

# Total of 934 images corresponds to 5 differnt classses are loaded in to this notebook 
                                                   #from the testing folder (harddrive)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential() # Sequentialy we are going to add the layers in this network

#Define the convolution layer with 16 kernals each of size 3x3
# This layer will accept the images of size 128x128x3
# Activation function for this layer is ReLU
model.add(Conv2D(16, (3, 3), input_shape = (128, 128, 3), activation = 'relu'))

#Maxpool layer generally defined at the end of each convolution layer. 
#Poolsize of 2x2 will bring down the feature map(convolution layer output) size by half
model.add(MaxPooling2D(pool_size = (2, 2)))

# Define the second convolution and maxpool layer
model.add(Conv2D(32, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

#Third Convolution and Maxpool Layer
model.add(Conv2D(64, (3, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))

# Flatten the feature map produced by the last convolution layer as single vector
model.add(Flatten())

#Add the hidden layers
model.add(Dense(units = 128,activation = 'relu'))

# Output layer should have 5 neurons as this data is having 5 classes of flowers
#"Softmax" activation function is preferred for Multiclass problem 
model.add(Dense(units = 5, activation = 'softmax'))
```

```python
#compile the CNN model with 'adam' optimizer to minimize the loss (cross entropy)
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Fit(Train) the model using trainset of data
# How many time the weight needs to be updated in one iteratin will be decided by steps_per_epoch
# How many iteration for which the training need to be happen will be decided by 'epochs'

#validation_steps determines how many batches of the validation data will be processed to evaluate the model's 
#performance at the end of each epoch.

#934 // 128 = 7 means that during validation, the model will process 7 batches of 128 images from the validation data.

model.fit(training_set,
                         steps_per_epoch =2736//128,  # 21.375.
                         epochs = 3,
                         validation_data = test_set,
                         validation_steps = 934//128)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image

# Define the number of layers to visualize (layer indices)
layer_outputs = [layer.output for layer in model.layers[1:8]]  # First 8 layers (adjust if needed)

print(layer_outputs)
```

```python
# Test image path
test_image = 'Training/daisy/21652746_cc379e0eea_m.jpg'

# Load the test image and preprocess it
img = image.load_img(test_image, target_size=(128, 128))  # Resize image to match the input shape
img_arr = image.img_to_array(img)
img_arr = np.expand_dims(img_arr, axis=0)  # Add batch dimension
img_arr /= 255.  # Normalize image
```

```python
# Create a new model to extract activations for the selected layers
activation_model = Model(inputs=model.input, outputs=layer_outputs)
activations = activation_model.predict(img_arr)
```

```python
# Print the actual shape of the activations (1, height, width, channels)
for activation in activations:
    print(activation.shape) 
   
```

```python
# Select the activations for the specific layers
conv2d_1_activations = activations[0]  # First convolutional layer (conv2d_1)
maxpool_1_activations = activations[1]  # First pooling layer (maxpool2d_1)
conv2d_2_activations = activations[2]  # Second convolutional layer (conv2d_2)
maxpool_2_activations = activations[3]  # Second pooling layer (maxpool_2)
conv2d_3_activations = activations[4]  # Third convolutional layer (conv2d_3)
maxpool_3_activations = activations[5]  # Third pooling layer (maxpool_3)
```

```python
import matplotlib.pyplot as plt

# Helper function to visualize filters for a given layer
def visualize_filters(layer_activations, layer_name):
    fig = plt.figure(figsize=(20, 8))  # Increase figure width for a single row layout
    num_filters = layer_activations.shape[-1]  # Number of filters in the layer

    # Loop through all filters to visualize
    for img in range(num_filters):  # Display all feature maps in a single row
        ax = fig.add_subplot(1, num_filters, img+1)  # Create 1xN grid of subplots (1 row)
        ax.imshow(layer_activations[0, :, :, img], cmap='viridis')  # Display activation map for each filter
        plt.xticks([])  # Remove x-axis ticks
        plt.yticks([])  # Remove y-axis ticks
        #plt.title(f'Filter {img+1}', fontsize=10)  # Add title with filter number

    # Print the shape of one filter's activation map (since all filters have the same shape)
    print(f'Shape of each filter in {layer_name}: {layer_activations[0, :, :, 0].shape}')

    fig.subplots_adjust(wspace=0.4)  # Adjust spacing between subplots
    plt.show()

# Visualize conv2d_1 and maxpool_1 activations
visualize_filters(conv2d_1_activations, 'conv2d_1')
visualize_filters(maxpool_1_activations, 'maxpool_1')

# Visualize conv2d_2 and maxpool_2 activations
visualize_filters(conv2d_2_activations, 'conv2d_2')
visualize_filters(maxpool_2_activations, 'maxpool_2')

# Visualize conv2d_3 and maxpool_3 activations
visualize_filters(conv2d_3_activations, 'conv2d_3')
visualize_filters(maxpool_3_activations, 'maxpool_3')
```
