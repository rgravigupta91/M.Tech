# Market-1501 Dataset Overview

The **Market-1501** dataset is a widely-used benchmark for evaluating the performance of person re-identification systems. It was introduced to help develop and test algorithms that can identify or re-identify individuals across different camera captures within a closed-circuit television (CCTV) network, typically in public and semi-public areas such as markets, airports, and campuses.

## Key Characteristics

### Composition
- The dataset contains images from six different cameras.
- It includes a total of 32,668 annotated bounding boxes of 1,501 identities.
- Each identity is captured by at least two cameras, ensuring cross-camera views.

### Annotations
- Each image is annotated with a bounding box that locates the person within the frame.
- The dataset provides both hand-drawn and DPM-detected (Deformable Part Model) bounding boxes.

### Evaluation Protocol
- The Market-1501 dataset is usually split into two parts: a training set and a test set. The training set includes 751 identities, while the test set comprises the remaining 750 identities.
- Performance metrics commonly used include the Cumulative Matching Characteristic (CMC) curve and the mean Average Precision (mAP), which measure the accuracy of the identification and the precision of the detection, respectively.

## Usage in Research
This dataset has been fundamental in advancing the state-of-the-art in person re-identification, with numerous studies using it to benchmark algorithms that involve deep learning, feature extraction, and complex matching systems.

## Accessibility
The dataset is publicly available for research purposes, ensuring that researchers and developers can use it to train and test their person re-identification models.

# Create APN Dataset

```python
class APN_Dataset(Dataset):
    def __init__(self, df):
        self.df = df

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        A_img = io.imread(DATA_DIR + row.Anchor)
        P_img = io.imread(DATA_DIR + row.Positive)
        N_img = io.imread(DATA_DIR + row.Negative)

        A_img = torch.from_numpy(A_img).permute(2, 0, 1) / 255.0
        P_img = torch.from_numpy(P_img).permute(2, 0, 1) / 255.0
        N_img = torch.from_numpy(N_img).permute(2, 0, 1) / 255.0

        return A_img, P_img, N_img



```

# EfficientNet_B0 Model Overview

The **EfficientNet_B0** model is a part of the **EfficientNet family** of models, which are known for their efficiency and effectiveness in performing image classification tasks. The EfficientNet models were designed with a method called **compound scaling** that uniformly scales network width, depth, and resolution with a set of fixed scaling coefficients.

## Key Features of EfficientNet_B0

### Architecture
- **EfficientNet_B0** is the baseline model of the EfficientNet series. It is carefully balanced in terms of depth, width, and resolution, making it a good starting point for many image classification tasks.

### Performance
- It provides a good balance between accuracy and computational efficiency. It was designed using a neural architecture search that optimizes both accuracy and FLOPs (floating-point operations per second).

### Scalability
- The EfficientNet models can be easily scaled up (from B0 to B7) to achieve better accuracy when more computational power or higher accuracy is needed.


```python
class APN_Model(nn.Module):
    def __init__(self, emb_size=512):
        super(APN_Model, self).__init__()
        self.efficientnet = timm.create_model('efficientnet_b0', pretrained=True)
        self.efficientnet.classifier = nn.Linear(
            in_features=self.efficientnet.classifier.in_features,
            out_features=emb_size
        )

    def forward(self, images):
        embeddings = self.efficientnet(images)
        return embeddings
```

```python
model = APN_Model()
model.to(DEVICE)
```

# Create Train and Eval function

```python
def train_fn(model, dataloader, optimizer, criterion):
    model.train()  # Enable dropout layers, if any
    total_loss = 0.0

    for A, P, N in tqdm(dataloader):
        # Move data to the appropriate device (e.g., CPU or GPU)
        A, P, N = A.to(DEVICE), P.to(DEVICE), N.to(DEVICE)

        # Forward pass through the model to get embeddings
        A_embs = model(A)
        P_embs = model(P)
        N_embs = model(N)

        # Calculate the loss using the criterion (e.g., triplet loss)
        loss = criterion(A_embs, P_embs, N_embs)

        # Zero out gradients, backpropagate, and update weights
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Accumulate loss for reporting
        total_loss += loss.item()

    return total_loss / len(dataloader)
```

```python
def eval_fn(model, dataloader, criterion):
    model.eval()  # Disable dropout layers
    total_loss = 0.0

    with torch.no_grad():  # No gradient calculation required
        for A, P, N in tqdm(dataloader):
            A, P, N = A.to(DEVICE), P.to(DEVICE), N.to(DEVICE)

            # Forward pass through the model
            A_embs = model(A)
            P_embs = model(P)
            N_embs = model(N)

            # Calculate loss without updating model weights
            loss = criterion(A_embs, P_embs, N_embs)

            # Accumulate loss
            total_loss += loss.item()

    return total_loss / len(dataloader)

```

```python
criterion = nn.TripletMarginLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LR)
```

# Create Training Loop

```python
best_valid_loss = np.Inf

for i in range(EPOCHS):
    train_loss = train_fn(model, trainloader, optimizer, criterion)
    valid_loss = eval_fn(model, validloader, criterion)

    if valid_loss < best_valid_loss:
        torch.save(model.state_dict(), 'best_model.pt')
        best_valid_loss = valid_loss
        print('SAVED WEIGHTS SUCCESS')

    print(f"EPOCHS: {i + 1} train_loss: {train_loss} valid_loss: {valid_loss}")
```

# Get Anchor Embeddings

```python
def get_encoding_csv(model, anc_img_names):
    anc_img_names_arr = np.array(anc_img_names)
    encodings = []

    model.eval()
    with torch.no_grad():
        for i in tqdm(anc_img_names_arr):
            A = io.imread(DATA_DIR + i)
            A = torch.from_numpy(A).permute(2, 0, 1) / 255.0
            A = A.to(DEVICE)
            A_enc = model(A.unsqueeze(0))## (c, h, w) -> (1, c, h, w)
            encodings.append(A_enc.squeeze().cpu().detach().numpy())

    encodings = np.array(encodings)
    encodings = pd.DataFrame(encodings)
    df_enc = pd.concat([anc_img_names, encodings], axis=1)

    return df_enc
```

```python
def euclidean_dist(img_enc, anc_enc_arr):
    dist = np.sqrt(np.dot(img_enc - anc_enc_arr, (img_enc - anc_enc_arr).T))
    return dist
```
