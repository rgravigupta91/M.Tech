```python
import os
import yaml
import xml.etree.ElementTree as ET

# Define paths
input_folder = r"C:\Users\Milan Amrut Joshi\Desktop\Misc_2\DL\YOLO\annotations"
output_folder = r"C:\Users\Milan Amrut Joshi\Desktop\Misc_2\DL\YOLO\anno_yolo"
images_folder = r"C:\Users\Milan Amrut Joshi\Desktop\Misc_2\DL\YOLO\images"  # Path to images

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Set for storing class names (to avoid duplicates)
class_names = set()

# Function to convert XML to dictionary for YOLO-compatible YAML format
def xml_to_dict(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Base structure for the YAML data
    data = {
        'image': root.find('filename').text,
        'annotations': []
    }

    # Iterate over each object in the XML
    for obj in root.findall('object'):
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # Append bounding box data to annotations list
        data['annotations'].append({
            'class': name,
            'bbox': [xmin, ymin, xmax, ymax]
        })

        # Add class name to set
        class_names.add(name)

    return data

# Convert each XML file in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".xml"):
        xml_path = os.path.join(input_folder, filename)
        
        # Convert XML to dictionary
        data = xml_to_dict(xml_path)

        # Define the output YAML file path
        yaml_filename = os.path.splitext(filename)[0] + ".yaml"
        yaml_path = os.path.join(output_folder, yaml_filename)

        # Write data to YAML file
        with open(yaml_path, 'w') as yaml_file:
            yaml.dump(data, yaml_file, default_flow_style=False)

        print(f"Converted {filename} to {yaml_filename}")

# Prepare data.yaml content
data_yaml = {
    'train': os.path.join(images_folder, 'train'),  # Path to training images
    'val': os.path.join(images_folder, 'val'),      # Path to validation images
    'nc': len(class_names),                         # Number of classes
    'names': list(class_names)                      # List of class names
}

# Save data.yaml
data_yaml_path = os.path.join(output_folder, 'data.yaml')
with open(data_yaml_path, 'w') as yaml_file:
    yaml.dump(data_yaml, yaml_file, default_flow_style=False)

print(f"data.yaml created at {data_yaml_path}")
```
