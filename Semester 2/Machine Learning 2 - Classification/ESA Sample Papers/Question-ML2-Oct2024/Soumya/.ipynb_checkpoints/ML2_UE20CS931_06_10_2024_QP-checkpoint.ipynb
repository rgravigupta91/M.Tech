```python

X_train_rs,X_test_rs,y_train_rs,y_test_rs=train_test_split(x_resample,y_resample,test_size=0.2,random_state=42)
rfe_model=RandomForestClassifier(random_state=42)
rfe=RFE(rfe_model,n_features_to_select=5)
rfe.fit(X_train_rs,y_train_rs)
```

```python
selected_features=X_train_rs.columns[rfe.support_]
selected_features
```

```python
# hyperparamter tuning
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

# Define parameter grid
rf_param_grid={
    'n_estimators':[50,100,200,300],
    'max_depth':[5,10,15,None],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
    }



# Initialize model
rf_model = RandomForestClassifier(random_state=42)

# Perform RandomizedSearchCV
rf_random = RandomizedSearchCV(rf_model, rf_param_grid, n_iter=10, cv=3, scoring='recall', n_jobs=-1, random_state=42)
rf_random.fit(X_train_rs[selected_features], y_train_rs)

# Best parameters
best_rf_model = rf_random.best_estimator_
print("Best Hyperparameters for Random Forest:", rf_random.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression

lr_param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

lr_model = LogisticRegression(penalty='l1', random_state=42)

lr_random = RandomizedSearchCV(lr_model, lr_param_grid, n_iter=10, cv=3, scoring='recall', n_jobs=-1, random_state=42)
lr_random.fit(X_train_rs[selected_features], y_train_rs)

best_lr_model = lr_random.best_estimator_
print("Best Hyperparameters for Logistic Regression:", lr_random.best_params_)
```

```python
from xgboost import XGBClassifier

xgb_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0]
}

xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss')

xgb_random = RandomizedSearchCV(xgb_model, xgb_param_grid, n_iter=10, cv=3, scoring='recall', n_jobs=-1, random_state=42)
xgb_random.fit(X_train_rs[selected_features], y_train_rs)

best_xgb_model = xgb_random.best_estimator_
print("Best Hyperparameters for XGBoost:", xgb_random.best_params_)
```

### 3 (d) From a business perspective answer the following: (5 Marks)
- a. Which data will you choose Balanced or Imbalanced and why? (2.5 Marks)
- b. Based on the given problem statement, explain which metric should be considered from the confusion matrix to gauge the effectiveness of the model built. (2.5 marks)


From a business perspective, Balanced Data is the better choice.

 Reason:

In an imbalanced dataset, the model may be biased toward the majority class (no stroke cases), leading to high accuracy but poor recall for stroke detection.

In a healthcare setting, missing a stroke case (False Negative) can have serious consequences, potentially risking a patients life.

Balancing the dataset (e.g., using SMOTE) ensures the model learns patterns in both stroke and non-stroke cases, improving recall for minority cases.

 We prefer balanced data to avoid bias and improve the detection of stroke cases.

For stroke prediction, the most important metric from the confusion matrix is Recall (Sensitivity / True Positive Rate).

Why Recall?
Recall measures the models ability to correctly identify stroke cases (True Positives).
In a healthcare scenario, False Negatives (misclassifying a stroke patient as healthy) can be life-threatening.
A high recall ensures the model correctly detects more stroke cases, reducing the risk of missing critical diagnoses.
