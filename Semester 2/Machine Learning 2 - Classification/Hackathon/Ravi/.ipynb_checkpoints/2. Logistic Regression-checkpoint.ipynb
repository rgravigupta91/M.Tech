```python
train_file_path = 'Train_Dataset.csv'
test_file_path = 'Test_Dataset.csv'

train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)
```

```python
train_data.dropna(subset=['EmployeeID'], thresh=1,inplace=True)
```

```python
def scale(df):
    scaler = MinMaxScaler()
    df = scaler.fit_transform(df)
    return df
```

```python
def Encode(df):
    le = LabelEncoder()
    for col in df.select_dtypes(include=['object']):
        df[col] = le.fit_transform(df[col].astype(str))
    return df
```

```python
def Fillnas(df):
    df.fillna(df.median(numeric_only=True), inplace=True)
    return df
```

```python
#train_data['MaritalStatus'] = train_data['MaritalStatus'].apply(lambda x: x if  x != 'M' else 'Married')
def conv_gender(df):
    df['MaritalStatus'] = df['MaritalStatus'].apply(lambda x: x if  x != 'M' else 'Married')
    return df
```

```python
def drop_col(df):
    df.drop('SalaryHikelastYear',axis=1,inplace=True)
    df.drop('ESOPs',axis=1,inplace=True)
    return df
```

```python
def run_pipeline(df):
#    pipeline = df.pipe(conv_gender).pipe(Fillnas).pipe(Encode).pipe(drop_col)
    pipeline = df.pipe(Fillnas).pipe(Encode).pipe(drop_col).pipe(scale)
    pipeline
```

```python
run_pipeline(train_data)
```
