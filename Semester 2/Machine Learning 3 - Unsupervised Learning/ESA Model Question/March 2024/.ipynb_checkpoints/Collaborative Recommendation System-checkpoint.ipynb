```python
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split, cross_validate
from surprise import SVDpp
from surprise import accuracy

import pandas as pd
import numpy as np
```

```python
ratings = pd.read_csv("Book_ratings.csv")
reader = Reader(rating_scale=(1, 5))
```

```python
rating_data = Dataset.load_from_df(ratings[['book_id','user_id','rating']],reader)
train, test = train_test_split(rating_data,test_size=0.2,random_state=42)
```

```python
trainsetfull = rating_data.build_full_trainset()
print('Number of users: ', trainsetfull.n_users, '\n')
print('Number of items: ', trainsetfull.n_items, '\n')
```

```python
alg=SVDpp()
alg.fit(trainsetfull)
```

```python
pred=alg.test(test)
pred
```

```python
accuracy.rmse(pred)
```
