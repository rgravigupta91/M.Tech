```python
from langchain import LLMChain, PromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

```

```python
!huggingface-cli login
```

```python
#  Load the Llama 3 model and tokenizer
model_name = "meta-llama/Llama-3.2-1B"  # Replace with your desired Llama model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype="auto")
```

```python
# Create a Hugging Face pipeline
llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=256)
```

```python
# Integrate Llama 3 with LangChain
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```

```python
# Define a prompt template
template = """You are a helpful AI assistant. Here is the user input:
{input}

Respond as clearly and helpfully as possible."""
prompt = PromptTemplate(input_variables=["input"], template=template)
```

```python
# Create the LangChain pipeline
chatbot_chain = LLMChain(llm=llm, prompt=prompt)
```

```python
# Start a conversation loop
print("Chatbot: Hello! How can I assist you today?")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a great day!")
        break

    # Generate a response using the LangChain pipeline
    response = chatbot_chain.run(input=user_input)
    print(f"Chatbot: {response}")
```
