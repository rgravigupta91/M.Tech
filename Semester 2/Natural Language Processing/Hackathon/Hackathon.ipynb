```python
import nltk
import pandas as pd
import random
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.probability import FreqDist
from nltk.classify import NaiveBayesClassifier
import nltk.classify.util
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv('train_(1).csv')
```

```python
df.head()
```

```python
df.insert(1, 'Comments', df['Review_Title'] + ' ' + df['Review'], allow_duplicates=True)
```

```python
df.head()
```

```python
df.drop(['Review_Title','Review'], axis=1, inplace=True)
```

```python
import unicodedata

def remove_acented_chars(text):
    return unicodedata.normalize('NFKD', text).encode('ascii','ignore').decode('utf-8','ignore')

df['Comments'] = df['Comments'].apply(remove_acented_chars)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return ' '.join([word for word in tokens.split() if word.lower() not in stop_words])

df['Comments'] = df['Comments'].apply(remove_stopwords)
```

```python
import string
def translator(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Remove punctuation
df['Comments'] = df['Comments'].apply(translator)
```
