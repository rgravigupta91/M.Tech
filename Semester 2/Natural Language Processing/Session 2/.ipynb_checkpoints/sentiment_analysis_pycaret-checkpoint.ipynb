# Data cleanup

### The text data cannot be used directly to train a classification model. it needs to be converted to numeric representation. Before converting to numeric representation, the raw text needs to be cleaned up

# cleanup-Basic:

#the basic cleanup step to remove
#1) remove html tags
#2) remove non-alphabetic charecters
#3) convert to lower case
#4) remove stop words
# use beautifulsoup, regex, nltk package

```python
def cleanup_reviews(raw_review):
    # 1. Remove HTML tags
    review_text = BeautifulSoup(raw_review, 'lxml').get_text()

    # 2. Remove non-alphabetic charecters using regex
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)

    # 3. Convert to lower case, split into individual words
    words = letters_only.lower().split()

    # 4. Create set of stopwords
    stops = set(stopwords.words("english"))

    # 5. Remove stop words
    meaningful_words = [w for w in words if not w in stops]

    # 6. Join the words back into one string separated by space,
    # and return the result.
    return( " ".join( meaningful_words ))
```

```python
df_imdb['cleaned_review']=df_imdb.review.apply(cleanup_reviews)
```

#cleanup-Advanced: Remove high frequency words that are common to both positive and negetive review movies

```python
# this function removed the words present in 'stops' from the 'raw_text'
def remove_commonwords(raw_text, stops=""):
    words = raw_text.split()
    meaningful_words = [w for w in words if not w in stops]
    return( " ".join( meaningful_words ))
```

```python
# this function removed the words not present in 'retain' from the 'raw_text'
def retain_onlywords(raw_text, retain=""):
    words = raw_text.split()
    meaningful_words = [w for w in words if  w in retain]
    return( " ".join( meaningful_words ))
```

```python
# function to get the frequency distribution of words.
#returns a dictionary of words and word count, maximum and minimum frequncy
def get_word_counts(df_ref, column_name):
    flt_txt= [a.split() for a in df_ref[column_name]]
    counter = Counter(chain.from_iterable(flt_txt))
    max_frq=max(counter.values())
    min_frq=min(counter.values())
    return (max_frq,min_frq, counter)
```

```python
# function to downsize the number of words below upper frequency and above lower frequency
def downsize_words(counter, upper, lower):
    new_dict={}
    df_wordcount=pd.DataFrame(columns=['Words', 'Count'])
    for key, value in counter.items():
        if (value<=upper and value>=lower):
            new_dict[key] = value
    return (new_dict)
```

```python
# the above wordcloud shows a lot of frequent common words between the two class-labels
# eliminating some of them might help identify key words that help distingush between + and - reviews
com_freq_words=['one', 'even', 'character', 'seem', 'sceen','get', 'film','see','like', 'seen', 'look', 'thing','movie', 'time', 'show', 'good','first','made', 'story','much','would','also','well','people', 'scene', 'movies', 'really','make','think', 'ways', 'characters', 'could']
df_imdb['cleaned_review2']=df_imdb.cleaned_review.apply(remove_commonwords, stops=com_freq_words)
```

#Building predictive model


 ### AutoML using pycaret
 ### The cleanedup reviews from the previous steps are used to train the model

```python
df_imdb_new=df_imdb[['cleaned_review2','sentiment' ]]
df_imdb_new.sample(10)
```

```python
cnt_vec=CountVectorizer(max_features=1000)
```

```python
# To limit the number of features considered, the max_features is set to 1000.
#the impact of setting the max_features to a higher number on the model accuracy is found to be negligible but significant on computational time
vectorizer=TfidfVectorizer(max_features=1000)
```

```python
Xcv = vectorizer.fit_transform(df_imdb_new.cleaned_review2)
count_vect_df = pd.DataFrame(Xcv.todense(), columns=vectorizer.get_feature_names_out())
count_vect_df['sentiment']=df_imdb_new.sentiment
```

```python
count_vect_df.shape
```

```python
count_vect_df.sample(10)
```

```python
#split data to train and validation samples
#10% of the data is reserved for validation. the validation samples will be used for predictions by the finalized model
df_train=count_vect_df.sample(frac=0.9,random_state=50)
df_test=count_vect_df.drop(df_train.index)
```

```python
print ('The number of samples in the train set is {}'.format(df_train.shape[0]))
print ('The number of samples in the test set is {}'.format(df_test.shape[0]))
print ('The number of features in each sample is {}'.format(df_train.shape[1]-1))
```

```python
df_train.info()
```

```python
from pycaret.classification import setup,compare_models,create_model, tune_model, evaluate_model, predict_model, finalize_model, save_model
```

```python
#this step is to initializes the environment in pycaret and creates the transformation pipeline to prepare the data for modeling
py_clas=setup(data=df_train,target='sentiment',  train_size=0.8,session_id=50)
```

```python
# Train different models and test
# the NaiveBayes' model, LDA and QDA, decision tree and Gradient boost classifier are excluded since these models tend to take a lot of time and return lower accuracy
# five fold cross validation is used and the metrics are reported rounded off the three decimal places
best_model= compare_models(verbose=True,   fold=5, round=3, turbo=True, exclude=['nb','lda', 'qda','dt','gbc'])
```

# The above table shows top-10 models ranked based on the accuracy and AUC scores.
# Logistic Regression classifier is the top performing model with an accuracy of 85% and AUC of 0.93.
This model is selected for the further exploration

```python
#test if the model has high variance/ over fitting tendency
py_clf=create_model('lr')
```

## The accurcy of the model is 85.3% +/-1.8% at 95% confidence level.
## The variance in the model is not high.

```python
#details of the logistic regression classifier
py_clf
```

# The logistic regression model is hyper tuned to extract the best model

```python
tuned_clf=tune_model(py_clf, round=3)
```

```python
tuned_clf
```

## The average accuracies of the base LR model and the tuned LR model are same but the variance for the tuned model is lower (0.91% in the base model to 0.8% in the tuned model)

```python
evaluate_model(tuned_clf)
```

```python
# predictions from the  tuned model on the test samples. Note this is the on the test samples heldout by pycaret not the df_val samples
predict_model(tuned_clf)
```

```python
final_clf = finalize_model(tuned_clf)
```
