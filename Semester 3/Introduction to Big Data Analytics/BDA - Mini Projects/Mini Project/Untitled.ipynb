```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 10))
ax.axis('off')

# Define components and positions (x, y, width, height)
components = [
    ("User Activity Logs\n(JSON Files)", (0.05, 0.75, 0.2, 0.1)),
    ("Kafka / Kinesis\n(Data Ingestion)", (0.3, 0.75, 0.2, 0.1)),
    ("Real-time Processing\n(Flink / Spark Streaming)", (0.55, 0.75, 0.3, 0.1)),
    ("Raw Data Storage\n(S3 / Cloud Storage)", (0.05, 0.5, 0.2, 0.1)),
    ("Processed Data (Real-time)\n(HBase / DynamoDB)", (0.4, 0.6, 0.25, 0.1)),
    ("Batch Processing\n(Spark / Dataflow)", (0.35, 0.4, 0.3, 0.1)),
    ("Processed Data (Batch)\n(Parquet / S3 / DWH)", (0.7, 0.5, 0.25, 0.1)),
    ("Analytics / Dashboards\n(Tableau / Superset)", (0.55, 0.25, 0.3, 0.1)),
]

# Draw components as rectangles
for text, (x, y, w, h) in components:
    ax.add_patch(mpatches.FancyBboxPatch((x, y), w, h,
                                         boxstyle="round,pad=0.02",
                                         edgecolor="black",
                                         facecolor="lightblue"))
    ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=10, weight='bold')

# Define arrows between components
arrows = [
    ((0.25, 0.8), (0.3, 0.8)),    # Logs → Kafka
    ((0.5, 0.8), (0.55, 0.8)),    # Kafka → Real-time Processing
    ((0.75, 0.8), (0.75, 0.7)),   # Real-time → Processed (Real-time)
    ((0.75, 0.7), (0.85, 0.55)),  # Processed (Real-time) → Batch Processed (DWH)
    ((0.75, 0.7), (0.75, 0.35)),  # Processed (Real-time) → Dashboard
    ((0.55, 0.75), (0.2, 0.55)),  # Real-time → Raw Storage
    ((0.2, 0.55), (0.35, 0.45)),  # Raw Storage → Batch Processing
    ((0.65, 0.45), (0.85, 0.55)), # Batch → Processed (Batch)
    ((0.85, 0.55), (0.75, 0.35)), # Batch Processed → Dashboards
]

# Draw arrows
for start, end in arrows:
    ax.annotate('', xy=end, xytext=start,
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))

plt.title("Big Data Architecture for Campaign Analytics", fontsize=14, weight='bold')
plt.tight_layout()
plt.show()
```
