#### Q3. Show horizontally One-third of the image in Red next one-third in blue and the last one third in green

```python
alphabet = io.imread('S2_1.png')
third = alphabet.shape[1] // 3
alphabet[:, :third, 1:] = 0
alphabet[:, third:2*third, [0, 2]] = 0
alphabet[:, 2*third:, :2] = 0

plt.imshow(alphabet)
plt.axis("off")
plt.show()
```
