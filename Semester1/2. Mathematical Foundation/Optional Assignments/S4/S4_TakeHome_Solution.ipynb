## Question 1 : PCA

### 1.A. Read the data S3TH_1.

```python
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
data = pd.read_csv("C:/Users/SUTITHI/Downloads/S3TH_1.csv")
data.head()
```

### 1.B. Print out the number of features in the dataset.

```python
data.shape
```

### 1.C. Conduct the PCA step by step for the given dataset

### 1.C.1 define a matrix with the dataframe

```python
data = np.array(data)
```

### 1.C.2 calculate the mean of each column

```python
M = mean(data.T, axis=1)
print(M)
```

### 1.C.3 center columns by subtracting column means

```python
C = data - M
print(C)
```

### 1.C.4 Find the covariance between the features

```python
V = cov(C.T)
print(V)
```

### 1.C.5 eigendecomposition (Eigen value and Eigen vector) of covariance matrix

```python
values, vectors = eig(V)
print(vectors)
print(values)
```

### 1.C.6 project data

```python
P = vectors.T.dot(C.T)
print(P.T)
```

### 1.C.7 create the PCA instance

```python
pca = PCA(n_components=5)
```

### 1.C.8 fit on data

```python
pca.fit(data)
```

### 1.C.9 access values and vectors

```python
print(pca.components_)
print(pca.explained_variance_)
```

### 1.C.10 transform data

```python
data_pca = pca.transform(data)
print(data_pca)
```

### 1.C.11 Plot the variance explained by each PCA component

```python
plt.plot(np.arange(1,6),pca.explained_variance_)
plt.show()
```

```python
plt.plot(np.arange(1,6),pca.explained_variance_ratio_)
plt.show()
```

### 5.A. Read the data S3TH_3.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skimage.io as io
from skimage.color import rgb2gray
```

```python
# read image in grayscale
img = io.imread('C:/Users/SUTITHI/Downloads/S3TH_3.png', as_gray=True)
plt.imshow(img)
plt.show()
```

### 5.B. Print out the number of features in the dataset.

```python
img.shape
```

### 5.C. Reduce the number of feature by 5% each time and show the visualise effect on the picture.

```python
# obtain svd
U, S, V = np.linalg.svd(img)
```

```python
# inspect shapes of the matrices
print(U.shape, S.shape, V.shape)
```

```python
len(S)
```

```python
# plot images with different number of components
index = np.arange(100,0,-5)
comp = index*len(S)/100
comps = list(map(lambda x:int(round(x)), comp))

plt.figure(figsize = (20, 15))
for i in range(len(comps)):
  low_rank = U[:, :comps[i]] @ np.diag(S[:comps[i]]) @ V[:comps[i], :]
  if(i  == 0):
     plt.subplot(4, 5, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("Original Image with n_components =" + str(comps[i]))
  else:
     plt.subplot(4, 5, i+1), plt.imshow(low_rank, cmap = 'gray'), plt.axis('off'), plt.title("n_components =" + str(comps[i]))
```
