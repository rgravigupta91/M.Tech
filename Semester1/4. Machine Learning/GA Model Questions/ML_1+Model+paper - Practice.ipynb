```python
def frequency_chart(attr,df):
    plt.figure(figsize=(5,5))
    #plt.subplot(1,2,1)
    plt1 = df[attr].value_counts().plot(kind='bar')
    plt.title('Frequency Distribution for '+attr)
    plt1.set(xlabel = attr, ylabel = 'Frequency')

    plt.show()
```

```python
def figure(attr,df):
    plt.figure(figsize=(10,5))
    
    plt.subplot(1,2,1)
    plt.title(attr+' Histogram')
    sns.countplot(x=df[attr],  palette=('cubehelix'))

    plt.subplot(1,2,2)
    plt.title(attr+' vs. Price')
    sns.boxenplot(x=df[attr], y = df.price, palette=("plasma"))

    plt.show()
               
```
